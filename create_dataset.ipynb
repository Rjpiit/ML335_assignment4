```python
import os
import requests
from bs4 import BeautifulSoup
# Define Google Image search URL
google_image = "https://www.google.com/search?tbm=isch&"

# Get user input for search term and number of images
data = input('What are you looking for? ')
n_images = int(input('How many images do you want? '))

# Calculate number of pages to search (each page typically contains 25 images)
n_pages = n_images // 20 + 1

# Initialize list to store image URLs
image_urls = []

# Iterate through each page of search results
for page in range(n_pages):
    start_index = page * 20
    search_url = google_image + 'q=' + data + '&start=' + str(start_index)

    # Send GET request to Google Images
    response = requests.get(search_url)

    # Parse HTML response using BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find all image elements in the page
    img_elements = soup.find_all('img')

    # Extract image URLs and add them to the list
    for img in img_elements:
        if img.get('src'):
            image_urls.append(img.get('src'))

# Filter out URLs that do not start with "https"
image_urls = [url for url in image_urls if url.startswith('https')]

# Print the extracted image URLs
for url in image_urls[:n_images]:
    print("i", url)
```

```python
i=0
saved_folder = data
if not os.path.exists(saved_folder):
    os.mkdir(saved_folder)
    print('Folder created:', saved_folder)

for link in image_urls[:n_images]:
    response = requests.get(link)

    image_name = saved_folder + '/' + data + str(i+1) + '.jpg'

    with open(image_name, 'wb') as fh:
        fh.write(response.content)
    print('Image downloaded:', image_name)
    i+=1
```
